```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

資料前處理
資料分為兩部分，一個是關鍵點坐標，總共 15 個關鍵點
坐標相關的值有 15 x 2 = 30 維； 另外還有 96 x 96 的灰階像素值。

讀取資料後，先簡單地過濾有缺失某些關鍵點坐標的資料，然後再把資料 scale 到 [-0.5, 0.5] 的範圍

```python
# 讀取 csv 文件
data = pd.read_csv("../data/face_training.csv")
```

```python
# 前 5 筆資料, .T 的作用是轉置，如果不理解可以和 data.head() 的結果相比較
data.head().T
```

```python
# ['欄位名稱']['第幾筆']
data['Image'][0]   
```

用 pandas 讀取 csv 文件後可以發現前 30 個 column 是坐標值，最後一個 column 是圖片像素值的字串 (以空白鍵區隔)，
接下來做資料的讀取以及前處理

```python
# 過濾有缺失值的 row
data = data.dropna()

# 將圖片像素值讀取為 numpy array 的形態
data['Image'] = data['Image'].apply(lambda img: np.fromstring(img, sep=' ')).values 

# 單獨把圖像 array 抽取出來
imgs = np.vstack(data['Image'].values)/255

# reshape 為 96 x 96
imgs = imgs.reshape(data.shape[0], 96, 96)

# 轉換為 float
imgs = imgs.astype(np.float32)

# 提取坐標的部分 (不包含最後一行,為像素)
points = data[data.columns[:-1]].values

# 轉換為 float
points = points.astype(np.float32)

# normalize 坐標值到 [-0.5, 0.5]
points = points/96 - 0.5
```

```python
print("圖像資料:", imgs.shape, "\n關鍵點資料:", points.shape)
```

```python
sample_img = imgs[1]   # 第2筆資料
sample_points = points[1]
```

```python
plt.imshow(sample_img, cmap='gray')
```

```python
print(sample_points)
```

```python
def plot_keypoints(img, points):
    # your code here
    # 顯示圖片
    plt.imshow(img, cmap='gray')
    # 把關鍵點畫上去
    for i in range(0,30,2):
        plt.scatter((points[i] + 0.5)*96, (points[i+1]+0.5)*96, color='red')
```

```python
plot_keypoints(sample_img, sample_points)
```
