```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('data/dog_bicycle.jpg')
h, w , _ = img.shape
```

```python
def show(img):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
show(img)
```

```python
boxes = np.array([[128, 224, 314, 537], [475, 85, 689, 170], [162, 119, 565, 441]]).astype(float)
boxes[:, [0, 2]] = boxes[:, [0, 2]] / img.shape[1]    # x1, x2
boxes[:, [1, 3]] = boxes[:, [1, 3]] / img.shape[0]    # y1, y2

img_show = img.copy()
for x1, y1, x2, y2 in boxes:
    cv2.rectangle(img_show, (int(x1*w), int(y1*h)), (int(x2*w), int(y2*h)), (0,255,0), 2)
show(img_show)
```

 VOC ,  20 ,  class index  1, 7, 16

```python
labels = np.array([1,7,6])
```

 bboxes  YOLO  tensor

```python
grid_num = 7
target = np.zeros((grid_num, grid_num, 30))
print("YOLO  tensor shape: ", target.shape)
```

```python
# 
cell_size = 1./grid_num # 
wh = boxes[:,2:] - boxes[:,:2] # bboxes  width  height  (x2,y2)-(x1,y1)
cxcy = (boxes[:,2:] + boxes[:,:2]) / 2 # bboxes 

for i in range(len(boxes)):
    cxcy_sample = cxcy[i]    
    ij = np.ceil((cxcy_sample/cell_size))-1 # bboxes  index
    target[int(ij[1]), int(ij[0]), 4] = 1 #  box  confidence  
    target[int(ij[1]), int(ij[0]), 9] = 1 #  box  confidence 
    target[int(ij[1]), int(ij[0]), int(labels[i])+9] = 1 # 
    xy = ij*cell_size # 
    
    #  x,y,w,h
    delta_xy = (cxcy_sample -xy)/cell_size
    target[int(ij[1]), int(ij[0]), 2:4] = wh[i]
    target[int(ij[1]), int(ij[0]), :2] = delta_xy
    target[int(ij[1]), int(ij[0]), 7:9] = wh[i]
    target[int(ij[1]), int(ij[0]), 5:7] = delta_xy
    
print(" 7x7x30  box  confidence \n", target[:, :, 4])
```

 tensor ,  1 ,  0 

 7x7  tensor  1 

```python
for h in np.arange(0, img.shape[0], img.shape[0]/grid_num).astype(int):
    cv2.line(img_show, (0, h), (img.shape[1], h), (0, 0, 0), 2)
for w in np.arange(0, img.shape[1], img.shape[1]/grid_num).astype(int):
    cv2.line(img_show, (w, 0), (w, img.shape[0]), (0, 0, 0), 2)
    
show(img_show)
```
